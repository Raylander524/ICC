### 19. Mensagens  idênticas  compartilhadas  pelo  mesmo  usuário  (e  suas quantidades);

```python
# Conectando ao banco existente
con = duckdb.connect('meu_banco.duckdb')

# Identifica mensagens repetidas por usuário
mensagens_repetidas = con.execute("""
    SELECT 
      id_member_anonymous, 
      text_content_anonymous,
      COUNT(*) AS total_repeticoes
    FROM telegram
    WHERE text_content_anonymous IS NOT NULL 
      AND id_member_anonymous IS NOT NULL
    GROUP BY id_member_anonymous, text_content_anonymous
    HAVING COUNT(*) > 1
    ORDER BY total_repeticoes DESC;
""").fetchdf()

# Fechando a conexão
con.close()

# Printando o resultado
print(mensagens_repetidas)
```

### 20. Mensagens  idênticas  compartilhadas  pelo  mesmo  usuário  em  grupos distintos (e suas quantidades);

```python
# Conectando ao banco existente
con = duckdb.connect('meu_banco.duckdb')

# Indetifica mensagens repetidas por grupo
repetidas_em_grupos = con.execute("""
    SELECT 
      id_member_anonymous, 
      text_content_anonymous,
      COUNT(DISTINCT id_group_anonymous) AS grupos_distintos
    FROM telegram
    WHERE text_content_anonymous IS NOT NULL
      AND id_member_anonymous IS NOT NULL

    GROUP BY id_member_anonymous, text_content_anonymous
    HAVING COUNT(DISTINCT id_group_anonymous) > 1
    ORDER BY grupos_distintos DESC;
""").fetchdf()

# Fechando a conexão
con.close()

# Printando o resultado
print(repetidas_em_grupos)
```

### 21. Os 30 unigramas, bigramas e trigramas mais compartilhados;

```python
!pip install scikit-learn
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Conectando ao banco existente
con = duckdb.connect('meu_banco.duckdb')

# Indetifica mensagens repetidas por grupo
textos = con.execute("""
    SELECT text_content_anonymous FROM telegram WHERE text_content_anonymous IS NOT NULL
""").fetchdf()

# Fechando a conexão
con.close()

# Função auxiliar para extrair os top n n-gramas
def top_ngrams(texts, ngram_range=(1,1), top_n=30):
    vectorizer = CountVectorizer(ngram_range=ngram_range, lowercase=True)
    X = vectorizer.fit_transform(texts)
    total_counts = X.sum(axis=0).A1
    vocab = vectorizer.get_feature_names_out()
    counts = pd.Series(total_counts, index=vocab)
    return counts.sort_values(ascending=False).head(top_n)

# Unigramas
top_unigrams = top_ngrams(textos['text_content_anonymous'], ngram_range=(1,1))
print("Top 30 Unigramas:\n", top_unigrams)

# Bigramas
top_bigrams = top_ngrams(textos['text_content_anonymous'], ngram_range=(2,2))
print("\nTop 30 Bigramas:\n", top_bigrams)

# Trigramas
top_trigrams = top_ngrams(textos['text_content_anonymous'], ngram_range=(3,3))
print("\nTop 30 Trigramas:\n", top_trigrams)
```

### 24. O usuário mais otimista;

### 28. O dia em que foi publicado a maior quantidade de mensagens;

```python
# Conectar ao banco de dados
con = duckdb.connect("meu_banco.duckdb")

# Consulta para encontrar o dia com maior número de mensagens
dia_com_mais_mensagens = con.execute("""
SELECT 
     CAST(date_message AS DATE) AS dia, 
    COUNT(*) AS num_mensagens
FROM telegram
WHERE date_message IS NOT NULL
GROUP BY dia
ORDER BY num_mensagens DESC
LIMIT 1
""").fetchdf()

# Fechando a conexão
con.close()

# Printando o resultado
print("Dia com maior número de mensagens:\n", dia_com_mais_mensagens)
```

### 29. As mensagens que possuem as palavras “FACÇÃO” e “CRIMINOSA”; 

```python
# Conectar ao banco de dados
con = duckdb.connect("meu_banco.duckdb")

# Consulta para encontrar as mensagens contendo as palavras "FACÇÃO" e "CRIMINOSA"
mensagens_faccao_criminosa = con.execute("""
SELECT text_content_anonymous
FROM telegram
WHERE text_content_anonymous IS NOT NULL
AND LOWER(text_content_anonymous) LIKE '%facção%' 
AND LOWER(text_content_anonymous) LIKE '%criminosa%'
""").fetchdf()

# Fechando a conexão
con.close()

# Printando o resultado
print("Mensagens com as palavras 'FACÇÃO' e 'CRIMINOSA':\n", mensagens_faccao_criminosa)
```

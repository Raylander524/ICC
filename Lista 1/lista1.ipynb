### Código com multithread

```python
# Inicializa o array de inconsistências.
array_inconsistencias = []

# Define a função de identificar as inconsistências.
def f(coluns):
    # Identifica o tipo da coluna.
    type_colun = type(df[coluns].dtype)
    # Instância um array interno para a utilização da multithread.
    array_inconsistencias = []
    # Loop para percorrer todas as linhas da coluna.
    for i in range(len(df)):
        # Verifica se o tipo do dado é diferente do tipo da coluna.
        if  isinstance(df.at[i, coluns], type_colun) :
            # Se o tipo do dados for diferente adicina ao array.
            array_inconsistencias.append((i, coluns))
    # Retorna o array interno.
    return array_inconsistencias

# Conta o número de cpus possíveis para thread.
num_processos = multiprocessing.cpu_count()

# Utilizando a biblioteca multiprocessing para executar o código em multiplas threads, onde é passado a quantidade de threads possíveis que pode abrir.
with multiprocessing.Pool(processes=num_processos) as pool:
    # Adiciona ao array global o resultado obtido em cada array analisado em cada thread.

    array_inconsistencias.append(pool.map(f, df.columns))
```

```python
# Printa o array
array_inconsistencias
```

## k)  Crie uma coluna chamada “sentiment” contendo os valores: -1 para textos negativos, 0 para textos neutros e 1 para textos positivos.

### Analisando sentimento através da coluna score_sentiment

```python
# Cria a função que atribui -1 para sentimentos negativos, 0 para sentimento neutros e 1 para sentimentos positivos
def sentimento(val):
    # Verifica se o valor não é nulo
    if pd.isna(val):
        # Se for nulo retorna 0
        return 0   
    # Verifica se o sentimento é negativo
    elif val < 0:
        # Se for negativo retorna -1
        return -1
    # Verifica se o sentimento é 0
    elif val == 0:
        # Se for retorna 0
        return 0
    # Caso nenhuma das opções anteriores seja atendida é porque o sentimento é positivo
    else:
        # Entrando no else é retornado 1 que é sentimento positivo
        return 1

# Criando a coluna 'sentiment' onde é feito a análise da coluna 'scorte_sentiment' que retorna o valor do sentimento, com valores entre -1 e 1
df['sentiment'] = df['score_sentiment'].map(lambda x: sentimento(x))
# Printa a coluna sentiment 
df['sentiment']
```

## l)  Eliminar as linhas cujo valor da coluna “text” contenham “trava-zaps”.

```python
# Importando a biblioteca de regex
import re

# Criando uma função que identifica trava-zaps
def verificar_trava_zap(texto):
    # 1. Verificar tamanho do texto, principal indicativo de um trava-zap
    if len(texto) > 10000:
        # 2. Caracteres invisíveis/suspeitos
        caracteres_invisiveis = ['\u200b', '\u200c', '\u200d', '\u2063', '\ufeff']
        if any(char in texto for char in caracteres_invisiveis):
            return True

        # 3. Repetição exagerada de emojis ou caracteres
        if re.search(r'(.)\1{30,}', texto):  # mesmo caractere repetido 30 vezes ou mais
            return True

        # 4. Mistura incomum de alfabetos (regex para identificar scripts diferentes)
        if re.search(r'[а-яА-Я]+.*[a-zA-Z]+|[a-zA-Z]+.*[а-яА-Я]+', texto):
            return True
    
    # Se não tiver nenhuma condição válida ele retorna false
    return False
```

```python
# Cria a coluna 'trava_zap_analisado' utilizando a função verifica_trava_zap
df['trava_zap_analisado']= df['text_content_anonymous'].apply(lambda x:verificar_trava_zap(x) if isinstance(x, str) else False)
# Printa a coluna trava_zap_analisado
df['trava_zap_analisado']
```

### Código com multithread

```python
array_inconsistencias = []

def f(coluns):
    type_colun = type(df[coluns].dtype)
    array_inconsistencias = []
    for i in range(len(df)):
        if  isinstance(df.at[i, coluns], type_colun) :
            #print(f"Na coluna {coluns} na linha {i} tem valor vazio")
            array_inconsistencias.append((i, coluns))
    return array_inconsistencias

num_processos = multiprocessing.cpu_count()

with multiprocessing.Pool(processes=num_processos) as pool:
    array_inconsistencias.append(pool.map(f, df.columns))
```

```python
array_inconsistencias
```

### Código com loop

```python
for coluns in df.columns:
    type_colun = type(df[coluns].dtype)
    for i in range(len(df)):
        if isinstance(df.at[i, coluns], type_colun):
            #print(f"Na coluna {coluns} na linha {i} tem valor vazio")
            array_inconsistencias.append((i, coluns))
```

## k)  Crie uma coluna chamada “sentiment” contendo os valores: -1 para textos negativos, 0 para textos neutros e 1 para textos positivos.

### Analisando sentimento através da coluna score_sentiment

```python
def sentimento(val):
    if pd.isna(val):
        return 0   
    elif val < 0:
        return -1
    elif val == 0:
        return 0
    else:
        return 1
df['sentiment'] = df['score_sentiment'].map(lambda x: sentimento(x))
df['sentiment']
```

### Analisando sentimento através da biblioteca LeIA

```python
from LeIA import SentimentIntensityAnalyzer

s = SentimentIntensityAnalyzer()

df['score_sentiment_leia'] = df['text_content_anonymous'].apply(lambda x: s.polarity_scores(x)['compound'] if isinstance(x, str) else 0)
df['score_sentiment_leia']
```

### Analisando sentimento através da biblioteca vader

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

s = SentimentIntensityAnalyzer()

df['score_sentiment_vader'] = df['text_content_anonymous'].apply(lambda x: s.polarity_scores(x)['compound'] if isinstance(x, str) else 0)
df['score_sentiment_vader']
```

### Diferença da analise de sentimento realizado com api para analise de sentimento de texto em português baseada na vader que é a LeIA feita a partir da coluna score_sentiment

```python
def sentimento(val):
    if pd.isna(val):
        return 0   
    elif val < 0:
        return -1
    elif val == 0:
        return 0
    else:
        return 1
df['sentiment_leia'] = df['score_sentiment_leia'].map(lambda x: sentimento(x))

diferencas = df['sentiment'] != df['sentiment_leia']

# Exibir os valores das colunas apenas nas linhas diferentes
valores_diferentes = df.loc[diferencas, ['sentiment', 'sentiment_leia']]

print(valores_diferentes)
```

### Diferença da analise de sentimento realizado com api para analise de sentimento chamada vader feita a partir da coluna score_sentiment

```python
def sentimento(val):
    if pd.isna(val):
        return 0   
    elif val < 0:
        return -1
    elif val == 0:
        return 0
    else:
        return 1
df['sentiment_vader'] = df['score_sentiment_vader'].map(lambda x: sentimento(x))

diferencas = df['sentiment'] != df['sentiment_vader']

# Exibir os valores das colunas apenas nas linhas diferentes
valores_diferentes = df.loc[diferencas, ['sentiment', 'sentiment_vader']]

print(valores_diferentes)
```

### A diferença nas três colunas de analises de sentimento

```python
diferencas = ((df['sentiment'] != df['sentiment_vader'])| (df['sentiment'] != df['sentiment_leia']))

# Exibir os valores das colunas apenas nas linhas diferentes
valores_diferentes = df.loc[diferencas, ['sentiment', 'sentiment_vader', 'sentiment_leia',]]

print(valores_diferentes)
```
